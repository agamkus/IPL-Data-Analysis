## Question 2:

### Loading "Deliveries" dataset:

```python
data2 = pd.read_csv("deliveries.csv")
data2.head()
```

### Replacing `Deccan Chargers` with `Sunrisers Hyderabad`:

```python
data2 = data2.replace("Deccan Chargers", "Sunrisers Hyderabad")
```

#### Solution: Filtering data with David Warner as batsman and calculating average per match strike rate:

```python
dw = data2.loc[data2['batsman'] == 'DA Warner']
dw_balls = pd.DataFrame(dw.groupby('match_id')['ball'].count())
dw_runs = pd.DataFrame(dw.groupby('match_id')['batsman_runs'].sum())
dw_data = pd.merge(dw_balls, dw_runs, on='match_id')
dw_data['Strike Rate'] = 100 * dw_data['batsman_runs']/dw_data['ball']
dw_data = dw_data.rename({'ball' : 'Balls', 'batsman_runs' : 'Runs'}, axis = 1)
dw_data
```

#### Getting mean of the Strike rate  match wise:

```python
avg_sr = dw_data['Strike Rate'].mean()
avg_sr
```

**Answer:** Average Strike Rate of David Warner in each match is 120.45.

## Question 3: What will be KL Rahul's contribution to team total?

### Filtering data where batsman is KLRahul and batting team is Kings XI Punjs:

```python
df = data2.loc[(data2['match_id'] >= 7894) & (data2['batting_team'] == 'Kings XI Punjab')]
kxip_runs = pd.DataFrame(df.groupby('match_id')['total_runs'].sum())
kxip_runs = kxip_runs.rename({'total_runs' : 'kxip Runs'}, axis = 1)
rahul_runs = data2.loc[(data2['match_id'] >= 7894) & (data2['batsman'] == 'KL Rahul')]
rahul_runs = pd.DataFrame(rahul_runs.groupby('match_id')['batsman_runs'].sum())
rahul_runs = rahul_runs.rename({'batsman_runs' : "Rahul' Runs"}, axis = 1)
kxip_data = pd.merge(kxip_runs, rahul_runs, on = 'match_id')
kxip_data["Rahul' Contribution"] = kxip_data["Rahul' Runs"]/kxip_data['kxip Runs'] * 100
kxip_data
```

### Average Rahul's contribution:

```python
kxip_data["Rahul' Contribution"].mean()
```

**Answer:** As per the above analysis, we see that Rahul's contributon is around **27%** to total Punjab's score.
