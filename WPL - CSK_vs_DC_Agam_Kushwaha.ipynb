### Importing Library:

```python
import pandas as pd
```

### Loading "Matches" Dataset:

```python
data1 = pd.read_csv("matches.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data1.head()
```

### Replacing `Delhi Daredevils` with `Delhi Capitals`:

```python
data1 = data1.replace("Delhi Daredevils", "Delhi Capitals")
```

### Filtering Delhi and CSK matches only:

```python
csk_dc = data1.loc[((data1['team1'] == 'Chennai Super Kings') & (data1['team2'] == 'Delhi Capitals') | (data1['team1'] == 'Delhi Capitals') & (data1['team2'] == 'Chennai Super Kings'))]
csk_dc.head(10)
```

## Question 1: Who will win the match?

#### Solution: Checking counts of winner:

```python
csk_dc['winner'].value_counts()/csk_dc.shape[0] * 100
```

**Answer:** As per the above analysis, winning percentage of CSK is **71.43** over DC. But, in this season, DC is performing well and their team combination is also nice. I would predict that DC will win today's match.

## Question 2: How many runs will CSK score in the match?

#### Solution: Observing runs of CSK in each match:

### Loading "Deliveries" dataset:

```python
data2 = pd.read_csv("deliveries.csv")
data2.head()
```

### Replacing `Delhi Daredevils` with `Delhi Capitals`:

```python
data2 = data2.replace("Delhi Daredevils", "Delhi Capitals")
```

### Filtering CSK and Delhi matches where batting team is CSK and bowling team is DC:

```python
csk_runs = data2.loc[(data2['batting_team'] == "Chennai Super Kings") & (data2['bowling_team'] == "Delhi Capitals")]
csk_runs = pd.DataFrame(csk_runs.groupby('match_id')['total_runs'].sum())
csk_runs
```

### Average CSK runs:

```python
avg_csk_runs = csk_runs['total_runs'].mean()
avg_csk_runs
```

**Answer:** As per the above analysis, average score of CSK is **~ 164**.

## Question 3: How many wickets will CSK take from the 91st ball to the 120th ball?

#### Solution: Observing match wise DC wickets between 91st and 120th ball i.e. data between 16th and 20th over:

### Filtering CSK and Delhi matches where batting team is DC and bowling team is CSK:

```python
dc_wickets = data2.loc[(data2['batting_team'] == "Delhi Capitals") & (data2['bowling_team'] == "Chennai Super Kings")]
dc_wickets.head()
```

### Filtering data between 16th and 120th over:

```python
dc_wickets = dc_wickets.loc[(dc_wickets['over'] >= 16) & (dc_wickets['over'] <= 20)]
dc_wickets = pd.DataFrame(dc_wickets.groupby('match_id')['player_dismissed'].count())
dc_wickets = dc_wickets.rename({'player_dismissed' : 'No. of Wickets'}, axis=1)
dc_wickets
```

### Average Wickets:

```python
avg_wic = dc_wickets['No. of Wickets'].mean()
avg_wic
```

**Answer:** On an average, CSK takes 2 wickets between 91st and 120th ball.

## Question 5: How many no-balls will be bowled in the match?

#### Solution: Filtering data with CSK and Delhi Matches:

```python
csk_dc_nb = data2.loc[((data2['batting_team'] == "Chennai Super Kings") & (data2['bowling_team'] == "Delhi Capitals") | (data2['batting_team'] == "Delhi Capitals") & (data2['bowling_team'] == "Chennai Super Kings"))]
csk_dc_nb = pd.DataFrame(csk_dc_nb.groupby('match_id')['noball_runs'].sum())
csk_dc_nb
```

#### Getting mean of the no-ball runs match wise:

```python
avg_nb = csk_dc_nb['noball_runs'].mean()
avg_nb
```

**Answer:** Average no-ball run in each match is very less 0.81. So, I would prdict that around 1-2 no-balls will be bowled in this match.
