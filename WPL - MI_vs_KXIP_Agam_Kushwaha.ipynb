### Importing Library:

```python
import pandas as pd
```

### Loading "Matches" Dataset:

```python
data1 = pd.read_csv("matches.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data1.head()
```

### Filtering MI and KXIP matches only:

```python
mi_kxip = data1.loc[((data1['team1'] == 'Mumbai Indians') & (data1['team2'] == 'Kings XI Punjab') | (data1['team1'] == 'Kings XI Punjab') & (data1['team2'] == 'Mumbai Indians'))]
mi_kxip.head()
```

## Question 1: Who will win the match?

#### Solution: Checking counts of winner:

```python
mi_kxip['winner'].value_counts()/mi_kxip.shape[0] * 100
```

**Answer:** As per the above counts, MI and KXIP won 54 % and 46% respectively. But, as per the teams' skill in this season,  I would predict that KXIP will win today's match.

## Question 2:  To what type of bowler will Quinton De Kock get out to?

#### Solution: Observing dismissal kind of Quinton De Kock:

### Loading "Deliveries" dataset:

```python
data2 = pd.read_csv("deliveries.csv")
data2.head()
```

### Filtering RCB and KXIP matches where batting team is KXIP and bowling team is RCB:

```python
qdc_out = data2.loc[(data2['batsman'] == "Q de Kock")]
qdc_out = pd.DataFrame(qdc_out.groupby('bowler')['player_dismissed'].count())
qdc_out.sort_values('player_dismissed', ascending=False).head(10)
```

**Answer:** If we notice overall Q De Kock's dismissals, we found that most of the times he got out on fast bowler.

## Question 3:  How many wickets will KXIP lose in the powerplay?

#### Solution: Filtering data where batting team is KXIP:

```python
kxip_wickets = data2.loc[(data2['batting_team'] == 'Kings XI Punjab') & ((data2['over']>=1) & (data2['over'] <=6))]
kxip_wickets = pd.DataFrame(kxip_wickets.groupby('match_id')['player_dismissed'].count())
kxip_wickets
```

### Average wickets in powerplay:

```python
avg_wic = kxip_wickets['player_dismissed'].mean()
avg_wic
```

**Answer:** On an average, KXIP loses 1-2 wickets in powerplay.

## Question 4:  How wil KL Rahul get out?

#### Filtering KL Rahul's batting data:

```python
kl_dismissal = data2.loc[(data2['batsman'] == "KL Rahul")]
kl_dismissal = pd.DataFrame(kl_dismissal.groupby('dismissal_kind')['player_dismissed'].count())
kl_dismissal
```

**Answer:** As per the analysis, we see that KL Rahul mostly hets out by **caught**.

## Question 5: How many wickets will fall in total during the match?

#### Solution: Filtering data with MI and KXIP Matches:

```python
mi_kxip_wickets = data2.loc[((data2['batting_team'] == "Mumbai Indians") & (data2['bowling_team'] == "Kings XI Punjab") | (data2['batting_team'] == "Kings XI Punjab") & (data2['bowling_team'] == "Mumbai Indians"))]
mi_kxip_wickets = pd.DataFrame(mi_kxip_wickets.groupby('match_id')['player_dismissed'].count())
mi_kxip_wickets = mi_kxip_wickets.rename({'player_dismissed' : 'No. of wickets'}, axis=1)
mi_kxip_wickets
```

### Average MI amd KXIP wickets:

```python
avg_mi_kxip_wic = mi_kxip_wickets['No. of wickets'].mean()
avg_mi_kxip_wic
```

**Answer:** Aper the above analysis, it is found that average wickets between MI and KXIP match is 12-13.
