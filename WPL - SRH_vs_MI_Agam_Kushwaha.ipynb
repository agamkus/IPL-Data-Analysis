### Importing Library:

```python
import pandas as pd
```

### Loading "Matches" Dataset:

```python
data1 = pd.read_csv("matches.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data2 = pd.read_csv('deliveries.csv')
data1.head()
```

### Replacing `Deccan Chargers` with `Sunrisers Hyderabad`:

```python
data1 = data1.replace("Deccan Chargers", "Sunrisers Hyderabad")
```

### Filtering SRH and CSK matches only:

```python
srh_mi = data1.loc[((data1['team1'] == 'Sunrisers Hyderabad') & (data1['team2'] == 'Mumbai Indians') | (data1['team1'] == 'Mumbai Indians') & (data1['team2'] == 'Sunrisers Hyderabad'))]
srh_mi.head()
```

### Question 1: Which team will win the match?

#### Solution: Checking counts of winner:

```python
srh_mi['winner'].value_counts()/srh_mi.shape[0] * 100
```

**Answer:** As per the above counts, MI won 54.16 % matches against SRH.

## Question 2: How many wides will be bowled in the match?

#### Solution: Filtering data with SRH and CSK Matches:

```python
srh_mi_wic = data2.loc[((data2['batting_team'] == "Sunrisers Hyderabad") & (data2['bowling_team'] == "Mumbai Indians") | (data2['batting_team'] == "Mumbai Indians") & (data2['bowling_team'] == "Sunrisers Hyderabad"))]
srh_mi_wic = pd.DataFrame(srh_mi_wic.groupby('match_id')['player_dismissed'].count())
srh_mi_wic
```

#### Getting mean of the wickets match wise:

```python
srh_mi_wic['player_dismissed'].mean()
```

**Answer:** Average wide runs in each match is 7.77. So, I would prdict that around 8 wides will be bowled in this match.

### Loading "Deliveries" dataset:

```python
data2 = pd.read_csv("deliveries.csv")
data2.head()
```

### Replacing `Deccan Chargers` with `Sunrisers Hyderabad`:

```python
data2 = data2.replace("Deccan Chargers", "Sunrisers Hyderabad")
```

### Filtering SRH and CSK matches where batting team is SRH and bowling team is CSK:

```python
srh_csk_bowl = data2.loc[(data2['batting_team'] == "Sunrisers Hyderabad") & (data2['bowling_team'] == "Chennai Super Kings")]
srh_csk_bowl.head(20)
```

## Question 3: How many runs will David Warner score?

#### Solution: Observing match wise Rohit Sharma's runs:

```python
david_runs = pd.DataFrame(srh_csk_bowl[srh_csk_bowl['batsman'] == 'DA Warner'].groupby('match_id')['batsman_runs'].sum())
david_runs
```

#### Finding average runs scored by Rohit Sharma in each match:

```python
david_runs['batsman_runs'].mean()
```

**Answer:** The average score of Rohit Sharma in each match is 52 runs. So, I would predict the same for this match.

## Question 4: How many wickets will CSK lose in the match?

#### Solution: Filterng data where batting team is CSK and bowling team is SRH:

```python
csk_wickets = data2.loc[(data2['batting_team'] == "Chennai Super Kings") & (data2['bowling_team'] == "Sunrisers Hyderabad")]
csk_wickets
```

#### Observing match wise Rabada's wicket counts:

```python
wickets_per_match = csk_wickets.groupby('match_id')['player_dismissed'].count()
wickets_per_match = pd.DataFrame(wickets_per_match)
wickets_per_match
```

#### Average wickets per match:

```python
avg_wic = wickets_per_match['player_dismissed'].mean()
avg_wic
```

**Answer:** Average wickets for CSK in each match is 4.16. I would predict that CSK will lose 4-5 wickets.
