## Question 4:  What will be difference in Eoin Morgan's overall strike rate and the strike rate of the last 7 balls he plays?

#### Filtering data for Eoin Morgan as batsman and calculating overall strike rate:

```python
morgan = data2.loc[(data2['batsman'] == 'EJG Morgan')]
morgan_balls = pd.DataFrame(morgan.groupby('match_id')['ball'].count())
morgan_runs = pd.DataFrame(morgan.groupby('match_id')['batsman_runs'].sum())
morgan_data = pd.merge(morgan_balls, morgan_runs, on='match_id')
morgan_data['Strike Rate'] = 100 * morgan_data['batsman_runs']/morgan_data['ball']
morgan_data = morgan_data.rename({'ball' : 'Balls', 'batsman_runs' : 'Runs'}, axis = 1)
morgan_data = morgan_data.rename({'Strike Rate': 'Overall SR'}, axis=1)
morgan_data
```

#### Filtering data for Eoin Morgan as batsman and the last 7 ball he plays:

```python
morgan2 = data2.loc[(data2['batsman'] == 'EJG Morgan') & ((data2['over'] == 18) & (data2['ball'] == 5) | ((data2['over'] >= 19) & (data2['over'] <= 20)))]
morgan2_balls = pd.DataFrame(morgan2.groupby('match_id')['ball'].count())
morgan2_runs = pd.DataFrame(morgan2.groupby('match_id')['batsman_runs'].sum())
morgan2_data = pd.merge(morgan2_balls, morgan2_runs, on='match_id')
morgan2_data['Strike Rate'] = 100 * morgan2_data['batsman_runs']/morgan2_data['ball']
morgan2_data = morgan2_data.rename({'ball' : 'Balls', 'batsman_runs' : 'Runs'}, axis = 1)
morgan2_data = morgan2_data.rename({'Strike Rate': 'Last 7 Ball SR'}, axis=1)
morgan2_data
```

#### Merging above 2 Dataframes:

```python
sr_diff = pd.merge(morgan_data, morgan2_data, on = 'match_id')
sr_diff = sr_diff[['Overall SR', 'Last 7 Ball SR']]
sr_diff['SR Difference'] = (sr_diff['Overall SR'] - sr_diff['Last 7 Ball SR'])
sr_diff
```

#### Average Strike Rate difference:

```python
diff = abs(sr_diff['Overall SR'].mean() - sr_diff['Last 7 Ball SR'].mean())
diff
```

**Answer:** So, we found that average difference between Morgan's Overall Strike Rate and Last 7 ball strike rate is around **39.50**.

## Question 5: How many runs will RCB score off KKR in the death?

#### Solution: Filtering data with RCB and KKR Matches where batting team is RCB and bowling team is KKR:

```python
rcb_runs = data2.loc[(data2['batting_team'] == "Royal Challengers Bangalore") & (data2['bowling_team'] == "Kolkata Knight Riders") & ((data2['over'] >= 16) & (data2['over'] <= 20))]
rcb_runs = pd.DataFrame(rcb_runs.groupby('match_id')['total_runs'].sum())
rcb_runs
```

#### Getting mean of the death over runs match wise:

```python
avg_runs = rcb_runs['total_runs'].mean()
avg_runs
```

**Answer:** Average death over runs in each match is 47.57.
