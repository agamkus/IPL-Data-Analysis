## Question 2: How many runs will RR score in first 6 overs?

### Loading "Deliveries" dataset:

```python
data2 = pd.read_csv("deliveries.csv")
data2.head()
```

#### Solution: Filtering data with RR as batting team and CSK as bowling team. Also getting data beween 1 and 6 overs only:

```python
rr_runs = data2.loc[(data2['batting_team'] == "Rajasthan Royals") & (data2['bowling_team'] == "Chennai Super Kings")]
rr_runs = rr_runs.loc[(rr_runs['over'] >= 1) & (rr_runs['over'] <= 6)]
rr_runs = pd.DataFrame(rr_runs.groupby('match_id')['total_runs'].sum())
rr_runs = rr_runs.rename({'total_runs' : 'Powerplay Runs'}, axis = 1)
rr_runs
```

#### Getting mean of the powerplay runs match wise:

```python
avg_pp_runs = rr_runs['Powerplay Runs'].mean()
avg_pp_runs
```

**Answer:** Average runs in each match in first 6 overs is **46** for RR.

## Question 3: What will Deepak Chahar's economy rate be?

#### Solution: Observing match wise Deepak Chahr's economy rate by dividing runs by over:

```python
# Filtering data for Deepk Chahar as bowler:
deepak_runs = data2.loc[data2['bowler'] == "DL Chahar"]
economy_rate = pd.DataFrame(deepak_runs.groupby('match_id')['total_runs'].sum())/4
economy_rate = economy_rate.rename({'total_runs' : 'economy_rate'}, axis = 1)
economy_rate
```

#### Finding average economy rate in each match:

```python
avg_er = economy_rate['economy_rate'].mean()
avg_er
```

**Answer:** As per the above analysis, average per match economy rate of Deepak Chahr is **6.79**.

## Question 4: How many wickets will Jofra Archer take in the death?

#### Solution: Filterng data where bowler is Jofra Archer and overs are 16-20 (Death Overs):

```python
ja_wickets = data2.loc[(data2['bowler'] == "J Archer") & ((data2['over'] >= 16) & (data2['over'] <= 20))]
ja_wickets = pd.DataFrame(ja_wickets.groupby('match_id')['player_dismissed'].count())
ja_wickets = ja_wickets.rename({'player_dismissed' : 'No. of Wickets'}, axis = 1)
ja_wickets
```

#### Observing average wickets taken by Jofra Archer in death overs:

```python
avg_wickets = ja_wickets['No. of Wickets'].mean()
avg_wickets
```

**Answer:** As per my analysis, average wickets taken by Jofra Archer in death overs are **0.7**, which means **0-1 wickets** .

## Question 5: What will Steve Smith's strike rate be during the match?

#### Solution: Filtering data with Steve Smith as batsman and calculating average per match strike rate:

```python
ss = data2.loc[data2['batsman'] == 'SPD Smith']
ss_balls = pd.DataFrame(ss.groupby('match_id')['ball'].count())
ss_runs = pd.DataFrame(ss.groupby('match_id')['batsman_runs'].sum())
ss_data = pd.merge(ss_balls, ss_runs, on='match_id')
ss_data['Strike Rate'] = 100 * ss_data['batsman_runs']/ss_data['ball']
ss_data = ss_data.rename({'ball' : 'Balls', 'batsman_runs' : 'Runs'}, axis = 1)
ss_data
```

### Average Strike rate per match:

```python
avg_sr = ss_data['Strike Rate'].mean()
avg_sr
```

**Answer:** As per the above analysis, we can see that on an average Steve Smith keeps strike rate of **124 runs** per match.
