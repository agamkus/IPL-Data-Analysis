### Importing Library:

```python
import pandas as pd
```

### Loading "Matches" Dataset:

```python
data1 = pd.read_csv("matches.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data1.head()
```

```python
data1.loc[data1['season'] == 2015].head()
```

### Replacing `Delhi Daredevils` with `Delhi Capitals` and `Deccan Chargers` with `Sunrisers hyderabad`:

```python
data1 = data1.replace("Delhi Daredevils", "Delhi Capitals")
data1 = data1.replace("Deccan Chargers", "Sunrisers Hyderabad")
```

### Filtering Delhi and KXIP matches only:

```python
srh_dc = data1.loc[((data1['team1'] == 'Sunrisers Hyderabad') & (data1['team2'] == 'Delhi Capitals') | (data1['team1'] == 'Delhi Capitals') & (data1['team2'] == 'Sunrisers Hyderabad'))]
srh_dc.head(10)
```

## Question 1: Who will win the match?

#### Solution: Checking counts of winner:

```python
srh_dc['winner'].value_counts()/srh_dc.shape[0] * 100
```

**Answer:** As per the above analysis, both the teams won 13-13 matches.

## Question 2: How many runs will be scored totally in the match?

### Loading "Deliveries" dataset:

```python
data2 = pd.read_csv("deliveries.csv")
data2.head()
```

### Replacing `Delhi Daredevils` with `Delhi Capitals` and `Deccan Chargers` with `Sunrisers hyderabad`:

```python
data2 = data2.replace("Delhi Daredevils", "Delhi Capitals")
data2 = data2.replace("Deccan Chargers", "Sunrisers Hyderabad")
```

### Filtering Data for SRH and DC match:

```python
srh_dc_runs = data2.loc[((data2['batting_team'] == "Sunrisers Hyderabad") & (data2['bowling_team'] == "Delhi Capitals") | (data2['batting_team'] == "Delhi Capitals") & (data2['bowling_team'] == "Sunrisers Hyderabad"))]
srh_dc_runs = pd.DataFrame(srh_dc_runs.groupby('match_id')['total_runs'].sum())
srh_dc_runs
```

### Average total runs:

```python
srh_dc_runs['total_runs'].mean()
```

**Answer:** As per the above analysis, average Strike Rate of KL Rahul in first 6 overs is **124**.

## Question 3: How many such dismissal will occur in the match where the batsman gets out caught?

```python
caught = data2.loc[((data2['batting_team'] == "Sunrisers Hyderabad") & (data2['bowling_team'] == "Delhi Capitals") | (data2['batting_team'] == "Delhi Capitals") & (data2['bowling_team'] == "Sunrisers Hyderabad")) & (data2['dismissal_kind'] == 'caught')]
caught = pd.DataFrame(caught.groupby('match_id')['dismissal_kind'].count())
caught
```

### Average no. of Caughts:

```python
caught['dismissal_kind'].mean()
```

**Answer:** As per the above analysis, we see that Shreyas Iyer's contributon is around **17%** to total Delhi's score.

## Question 5: How many extras will be bowled?

```python
extra = data2.loc[((data2['batting_team'] == "Sunrisers Hyderabad") & (data2['bowling_team'] == "Delhi Capitals") | (data2['batting_team'] == "Delhi Capitals") & (data2['bowling_team'] == "Sunrisers Hyderabad")) & (data2['extra_runs'] != 0)]
extra = pd.DataFrame(extra.groupby('match_id')['extra_runs'].sum())
extra
```
