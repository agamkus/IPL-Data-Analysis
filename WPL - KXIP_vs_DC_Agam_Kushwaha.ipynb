## Question 2: What will KL Rahul's strike rate be in first 6 overs?

#### Solution: Observing runs of KL Rahul in each match:

### Loading "Deliveries" dataset:

```python
data2 = pd.read_csv("deliveries.csv")
data2.head()
```

### Replacing `Delhi Daredevils` with `Delhi Capitals`:

```python
data2 = data2.replace("Delhi Daredevils", "Delhi Capitals")
```

### Filtering Data where batsman is KL Rahul and over is between 1 and 6:

```python
rahul = data2.loc[(data2['batsman'] == 'KL Rahul') & ((data2['over'] >= 1) & data2['over'] <= 6)]
rahul_balls = pd.DataFrame(rahul.groupby('match_id')['ball'].count())
rahul_runs = pd.DataFrame(rahul.groupby('match_id')['batsman_runs'].sum())
rahul_data = pd.merge(rahul_balls, rahul_runs, on='match_id')
rahul_data['Strike Rate'] = 100 * rahul_data['batsman_runs']/rahul_data['ball']
rahul_data = rahul_data.rename({'ball' : 'Balls', 'batsman_runs' : 'Runs'}, axis = 1)
rahul_data
```

### Average Strike Rate of KL Rahul in 1 to 6 overs:

```python
avg_rahul_sr = rahul_data['Strike Rate'].mean()
avg_rahul_sr
```

**Answer:** As per the above analysis, average Strike Rate of KL Rahul in first 6 overs is **124**.

## Question 3: What will be Shreya's Iyer's contribution to DC's total score?

#### Solution: Shreyas is playing from Delhi side since 2015; i.e. mtch_id 518. Observing match wise DC runs and Shreyas Iyer's runs:

```python
df = data2.loc[(data2['match_id'] >= 518) & (data2['batting_team'] == 'Delhi Capitals')]
dc_runs = pd.DataFrame(df.groupby('match_id')['total_runs'].sum())
dc_runs = dc_runs.rename({'total_runs' : 'DC Runs'}, axis = 1)
shreyas_runs = data2.loc[(data2['match_id'] >= 518) & (data2['batsman'] == 'SS Iyer')]
shreyas_runs = pd.DataFrame(shreyas_runs.groupby('match_id')['batsman_runs'].sum())
shreyas_runs = shreyas_runs.rename({'batsman_runs' : "Sheyas' Runs"}, axis = 1)
dc_data = pd.merge(dc_runs, shreyas_runs, on = 'match_id')
dc_data["Shreyas' Contribution"] = dc_data["Sheyas' Runs"]/dc_data['DC Runs'] * 100
dc_data
```

### Average Shreyas' contribution:

```python
dc_data["Shreyas' Contribution"].mean()
```

**Answer:** As per the above analysis, we see that Shreyas Iyer's contributon is around **17%** to total Delhi's score.

## Question 4: How many wickets will be taken by KXIP bowlers excluding that taken by Mohammed Shami, from the 31st ball to 120th ball?

#### Solution: Observing wicket  counts of KXIP bowlers excluding Mohammed Shami. Filtering the data between 6th and 20th Over.

```python
kxip_wickets = data2.loc[(data2['bowling_team'] == 'Kings XI Punjab') & (data2['bowler'] != 'Mohammed Shami') & ((data2['over'] >= 6) & (data2['over'] <= 20))]
kxip_wickets = pd.DataFrame(kxip_wickets.groupby('match_id')['player_dismissed'].count())
kxip_wickets = kxip_wickets.rename({'player_dismissed' : 'No. of Wickets'}, axis = 1)
kxip_wickets
```

### Average wickets:

```python
kxip_wickets['No. of Wickets'].mean()
```
