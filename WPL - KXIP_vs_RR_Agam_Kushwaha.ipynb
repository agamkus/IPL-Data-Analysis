### Importing Library:

```python
import pandas as pd
```

### Loading "Matches" Dataset:

```python
data1 = pd.read_csv("matches.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data1.head()
```

### Filtering RR and KXIP matches only:

```python
kxip_rr = data1.loc[((data1['team1'] == 'Kings XI Punjab') & (data1['team2'] == 'Rajasthan Royals') | (data1['team1'] == 'Rajasthan Royals') & (data1['team2'] == 'Kings XI Punjab'))]
kxip_rr.head(10)
```

## Question 1: Who will win the match?

```python
kxip_rr['winner'].value_counts()/kxip_rr.shape[0] * 100
```

## Question 2: 

#### Solution: Observing runs of KL Rahul in each match:

### Loading "Deliveries" dataset:

```python
data2 = pd.read_csv("deliveries.csv")
data2.head()
```

```python
rahul = data2.loc[(data2['batsman'] == 'KL Rahul') & ((data2['over'] >= 1) & data2['over'] <= 6)]
rahul_balls = pd.DataFrame(rahul.groupby('match_id')['ball'].count())
rahul_runs = pd.DataFrame(rahul.groupby('match_id')['batsman_runs'].sum())
rahul_data = pd.merge(rahul_balls, rahul_runs, on='match_id')
rahul_data['Strike Rate'] = 100 * rahul_data['batsman_runs']/rahul_data['ball']
rahul_data = rahul_data.rename({'ball' : 'Balls', 'batsman_runs' : 'Runs'}, axis = 1)
rahul_data
```

## Question 3: 

```python
df = data2.loc[(data2['match_id'] >= 518) & (data2['batting_team'] == 'Delhi Capitals')]
dc_runs = pd.DataFrame(df.groupby('match_id')['total_runs'].sum())
dc_runs = dc_runs.rename({'total_runs' : 'DC Runs'}, axis = 1)
shreyas_runs = data2.loc[(data2['match_id'] >= 518) & (data2['batsman'] == 'SS Iyer')]
shreyas_runs = pd.DataFrame(shreyas_runs.groupby('match_id')['batsman_runs'].sum())
shreyas_runs = shreyas_runs.rename({'batsman_runs' : "Sheyas' Runs"}, axis = 1)
dc_data = pd.merge(dc_runs, shreyas_runs, on = 'match_id')
dc_data["Shreyas' Contribution"] = dc_data["Sheyas' Runs"]/dc_data['DC Runs'] * 100
dc_data
```

## Question 4: 

```python
dot_balls = data2.loc[((data2['batting_team'] == "Kings XI Punjab") & (data2['bowling_team'] == "Kolkata Knight Riders") | (data2['batting_team'] == "Kolkata Knight Riders") & (data2['bowling_team'] == "Kings XI Punjab"))]
dot_balls = dot_balls.loc[dot_balls['total_runs'] ==0]
dot_balls = pd.DataFrame(dot_balls.groupby('match_id')['total_runs'].count()).sort_values(by='total_runs', ascending=False)
dot_balls
```

## Question 5: What will be the status of Chris Gayle's batting durin KXIP innings?

#### Solution: Filtering data with CSK and Delhi Matches:

```python
gayle = data2.loc[(data2['batsman'] == 'CH Gayle')]
gayle_balls = pd.DataFrame(gayle.groupby('match_id')['ball'].count())
gayle_runs = pd.DataFrame(gayle.groupby('match_id')['batsman_runs'].sum())
gayle_data = pd.merge(gayle_balls, gayle_runs, on='match_id')
gayle_data['Strike Rate'] = 100 * gayle_data['batsman_runs']/gayle_data['ball']
gayle_data = gayle_data.rename({'ball' : 'Balls', 'batsman_runs' : 'Runs'}, axis = 1)
gayle_data
```

#### Getting mean of the Chris Gayle's strike rate:

```python
gayle_data['Strike Rate'].mean()
```

**Answer:** Average strike rate of Chris Gayle is **127**.
