## Question 2: How many runs will be scored totally in the match?

### Loading "Deliveries" dataset:

```python
data2 = pd.read_csv("deliveries.csv")
data2.head()
```

```python
rcb_srh_runs = data2.loc[((data2['batting_team'] == "Sunrisers Hyderabad") & (data2['bowling_team'] == "Royal Challengers Bangalore") | (data2['batting_team'] == "Royal Challengers Bangalore") & (data2['bowling_team'] == "Sunrisers Hyderabad"))]
rcb_srh_runs = pd.DataFrame(rcb_srh_runs.groupby('match_id')['total_runs'].sum())
rcb_srh_runs
```

#### Getting mean total runs match wise:

```python
rcb_srh_runs['total_runs'].mean()
```

**Answer:** Average wide runs in each match is 7.77. So, I would prdict that around 8 wides will be bowled in this match.

## Question 3: 

```python
rcb_wic = data2.loc[(data2['batting_team'] == 'Royal Challengers Bangalore') & (data2['bowling_team'] == 'Sunrisers Hyderabad') & ((data2['over'] >=1) & (data2['over'] <= 6))]
rcb_wic = pd.DataFrame(rcb_wic.groupby('match_id')['dismissal_kind'].count())
rcb_wic = rcb_wic.rename({'dismissal_kind': 'RCB Wickets'}, axis=1)

srh_wic = data2.loc[(data2['batting_team'] == 'Sunrisers Hyderabad') & (data2['bowling_team'] == 'Royal Challengers Bangalore') & ((data2['over'] >=1) & (data2['over'] <= 6))]
srh_wic = pd.DataFrame(srh_wic.groupby('match_id')['dismissal_kind'].count())
srh_wic = srh_wic.rename({'dismissal_kind': 'SRH Wickets'}, axis=1)
wickets = pd.merge(rcb_wic, srh_wic, on='match_id')
wickets['Wicket Difference'] = abs(wickets['RCB Wickets'] - wickets['SRH Wickets'])
wickets
```

### Avg wicket difference:

```python
wickets['Wicket Difference'].mean()
```

## Question 5: How many dot balls will be bowled in the match?

```python
dot_balls = data2.loc[((data2['batting_team'] == "Royal Challengers Bangalore") & (data2['bowling_team'] == "Sunrisers Hyderabad") | (data2['batting_team'] == "Sunrisers Hyderabad") & (data2['bowling_team'] == "Royal Challengers Bangalore"))]
dot_balls = dot_balls.loc[dot_balls['total_runs'] == 0]
dot_balls = pd.DataFrame(dot_balls.groupby('match_id')['total_runs'].count())
dot_balls
```

#### Average dot balls per match:

```python
dot_balls['total_runs'].mean()
```
